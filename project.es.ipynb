#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds

poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
cinturon_sur = ds[ds["level5"].isin(poblaciones)]

poblaciones_calc = ["Getafe", "Alcorcón"]

for poblacion in poblaciones_calc:
    subset = cinturon_sur[cinturon_sur["level5"] == poblacion].copy()
    subset["pps"] = subset["price"] / subset["surface"]
    media_precio_per_sqm = subset["pps"].mean()
    print(f"{poblacion} - Precio medio por metro cuadrado: {media_precio_per_sqm} USD/m²")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds

poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
cinturon_sur = ds[ds["level5"].isin(poblaciones)]

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(14, 10), constrained_layout=True)

axs = axs.flatten()

for i, poblacion in enumerate(poblaciones):
    subset = cinturon_sur[cinturon_sur["level5"] == poblacion]
    axs[i].scatter(subset["surface"], subset["price"], alpha=0.7)
    axs[i].set_title(poblacion)
    axs[i].set_xlabel("Superficie (m²)")
    axs[i].set_ylabel("Precio (USD)")

plt.suptitle("Diagrama de Dispersión por Población")
plt.show()
```
